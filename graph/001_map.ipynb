```python
from graphviz import Digraph
from IPython.display import Image
import os

# Crear el mapa conceptual amb nom i orientació d'esquerra a dreta
mapa_conceptual = Digraph(comment='Aparell Locomotor', graph_attr={'rankdir': 'LR'})

# Node principal
mapa_conceptual.node('A', 'Aparell Locomotor', shape='box')

# Nodes de primer nivell: Funcions i Components
mapa_conceptual.node('B', 'Funcions', shape='ellipse')
mapa_conceptual.node('C', 'Components', shape='ellipse')

# Subnodes de Funcions
mapa_conceptual.node('E', 'Estructural', shape='box')
mapa_conceptual.node('F', 'Protectora', shape='box')
mapa_conceptual.node('G', 'Motora', shape='box')
mapa_conceptual.node('H', 'Hematopoètica', shape='box')
mapa_conceptual.node('I', 'Metabòlica', shape='box')

# Subnodes de Components
mapa_conceptual.node('J', 'Ossos', shape='box')
mapa_conceptual.node('K', 'Músculs', shape='box')
mapa_conceptual.node('L', 'Articulacions', shape='box')

# Subnodes d'explicacions per Funcions i Components (tot blanc)
mapa_conceptual.node('E1', 'Dóna forma i alineació al cos', shape='plaintext')
mapa_conceptual.node('F1', 'Protegeix òrgans com el cor, pulmons, cervell', shape='plaintext')
mapa_conceptual.node('G1', 'Permet el moviment', shape='plaintext')
mapa_conceptual.node('H1', 'Forma cèl·lules sanguínies a la medul·la òssia', shape='plaintext')
mapa_conceptual.node('I1', 'Reserva de minerals com el calci i el fòsfor', shape='plaintext')

mapa_conceptual.node('J1', 'Estructura rígida que dóna suport al cos', shape='plaintext')
mapa_conceptual.node('K1', 'Permeten la contracció per generar moviment', shape='plaintext')
mapa_conceptual.node('L1', 'Uneixen ossos i permeten el moviment entre ells', shape='plaintext')

# Relacions principals (fletxes des del mateix punt)
mapa_conceptual.edge('A', 'B', minlen='2',label="etiqueta linia")
mapa_conceptual.edge('A', 'C', minlen='2')

# Relacions de Funcions amb les explicacions
mapa_conceptual.edge('B', 'E', minlen='2')
mapa_conceptual.edge('E', 'E1', minlen='2')

mapa_conceptual.edge('B', 'F', minlen='2')
mapa_conceptual.edge('F', 'F1', minlen='2')

mapa_conceptual.edge('B', 'G', minlen='2')
mapa_conceptual.edge('G', 'G1', minlen='2')

mapa_conceptual.edge('B', 'H', minlen='2')
mapa_conceptual.edge('H', 'H1', minlen='2')

mapa_conceptual.edge('B', 'I', minlen='2')
mapa_conceptual.edge('I', 'I1', minlen='2')

# Relacions de Components amb les explicacions
mapa_conceptual.edge('C', 'J', minlen='2')
mapa_conceptual.edge('J', 'J1', minlen='2')

mapa_conceptual.edge('C', 'K', minlen='2')
mapa_conceptual.edge('K', 'K1', minlen='2')

mapa_conceptual.edge('C', 'L', minlen='2')
mapa_conceptual.edge('L', 'L1', minlen='2')

# Generar els fitxers PNG
output_path = 'mapa_conceptual_funcions_components_mateix_punt'  # Sense extensió
mapa_conceptual.render(output_path, format='png', cleanup=True)

# Comprovar si el fitxer existeix i mostrar-lo
png_path = f"{output_path}.png"
if os.path.exists(png_path):
    print(f"El fitxer {png_path} s'ha generat correctament.")
    display(Image(filename=png_path))
else:
    print(f"No s'ha trobat el fitxer {png_path}.")
```
